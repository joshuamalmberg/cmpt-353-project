```python
import numpy as np
import pandas as pd
import math
from scipy.fftpack import fft
from math import pow
import matplotlib.pyplot as plt
from scipy import signal
from scipy import stats
```

```python
data = pd.read_csv ('data/walk_leftleg/walk-leftleg-0.csv')
```

```python
plt.plot(data['time'],data['ax'],color='red')
plt.title('accelerations along x-axis with raw data')
plt.xlabel('time')
plt.ylabel('ax')
plt.grid()
```

```python
plt.plot(data['time'],data['ay'],color='green')
plt.title('accelerations along y-axis with raw data')
plt.xlabel('time')
plt.ylabel('ay')
plt.grid()
```

```python
plt.plot(data['time'],data['az'],color='blue')
plt.title('accelerations along z-axis with raw data')
plt.xlabel('time')
plt.ylabel('az')
plt.grid()
```

```python
data = data[data['time']>2]
data = data[data['time']<12]
plt.plot(data['time'],data['ax'],color='red')
plt.plot(data['time'],data['ay'],color='green')
plt.plot(data['time'],data['az'],color='blue')
plt.title('3D accelerations with raw data')
plt.xlabel('time')
plt.ylabel('acceleration')
```

```python
#calculate absolute acceleration 
def calculate_abs(x,y,z):
    acc=np.sqrt(x*x + y*y + z*z)
    return acc
data['acc']=calculate_abs(data['ax'],data['ay'],data['az'])
plt.plot(data['time'],data['acc'])
plt.title('Total accelerations')
plt.xlabel('time')
plt.ylabel('acceleration')
plt.grid() 
```

```python
#butterworth filter
def butterworth(data):
    b, a = signal.butter(8, 0.05, btype='lowpass', analog=False)
    low_passed = signal.filtfilt(b, a, data)
    return low_passed
data['acc'] = butterworth(data['acc'])
plt.plot(data['time'],data['acc'])
plt.grid()
```

```python
#fourier transformation
start_time = data['time'].values[0]
end_time = data['time'].values[-1]
N = data['time'].shape[0]
T=(end_time-start_time)/N
x=np.linspace(0.0,N*T,N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
plt.title('Fourier Transform on walking data 0')

plt.ylabel('frequency')
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
```
